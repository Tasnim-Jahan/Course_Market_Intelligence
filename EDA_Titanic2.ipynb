Now we can replace the missing value with average age. But we are not doing that, we make cluster of 1st class, 2nd class and 3rd class pessanger to get maximum accuraacy and then find the median of each class and then replace the missing value with the median.

```python
# Simple Imputer
def impute_age(cols): #we define a function of age and pclass
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24 #if age is null then for 1st class passenger age we will assign age 37,for 2nd class age=29,for 3rd class age=24
    else:
        return Age
```

```python
#here we will call our user define function, input_age, and execute it to replace the null value with their median value based on their class
titanic_df['Age'] = titanic_df[['Age', 'Pclass']].apply(impute_age, axis = 1)
```

```python
#as a result there is no misssing value left in age that is called preprocessing
#as we dealt with zero, this replacemet is permanent
titanic_df.isnull().sum()
```

```python
#inplace=true menas it will remove permanently
titanic_df.drop(['Cabin'], axis = 1, inplace=True)
```

```python
titanic_df.isnull().sum()
```

```python
# Dropna() - Drops entire rows of data if less rows are missing
# Drops embarked missing rows
# we have two rows with missing values(Embarked) dropna will erase these rows
titanic_df.dropna(inplace = True)
```

After cleaning, the above dataset found. Now we need to find out the dependent(y) and independent columns(x1,x2,....xn) and then we will divide this dataset into four parts: x-test, y-test, x-train, y-train. Most cases we consider 70% data for trainig and 30% for testing.
Infact we have a seperate test data here, that test data is not the actual test data as we don't have lavel of that test data,only kaggle has the label so we cannnot consider that data as a test data. When we don't have seperate test data we consider training set and test set.When we have seperate test data we consider training set and validation set.

```python
#we are dropping some of the columns that are meaningless i.e."PassengerId", "Name", "Sex", "Ticket", "Embarked"
#and "Survived",as it will be our y/output.
X = titanic_df.drop(["PassengerId", "Survived", "Name", "Sex", "Ticket", "Embarked"], axis = 1)
y = titanic_df['Survived']
```

```python
# split X and y into training and testing sets
from sklearn.model_selection import train_test_split #(sklearn->science kit learn)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=41)
```

```python
# Import the class: Logistic Regression
#from sklearn.linear model that is another package we can import model
from sklearn.linear_model import LogisticRegression
```

In linear regression class label is numeric not diagonal

```python
# instantiate the model (using the default parameters)
logreg = LogisticRegression()
```

```python
# then fit the model with data, we build the model with xtrain and ytrain
logreg.fit(X_train, y_train)
```

```python
# the we check the Accuracy of the model
logreg.score(X_test, y_test)
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix, classification_report
predictions = logreg.predict(X_test)
confusion_matrix(y_test, predictions)

sns.heatmap(confusion_matrix(y_test, predictions), annot = True, fmt='0.0f')
```

```python
print(classification_report(y_test, predictions))
```

```python
# Model Evaluation using Confusion Matrix
y_pred = logreg.predict(X_test)
```

```python
# import the metrics class
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
# Visualizing Confusion Matrix using Heatmap
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
# Confusion Matrix Evaluation Metrics
# Evaluate the model using model evaluation metrics such as accuracy, precision, and recall.
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
print("Precision:", metrics.precision_score(y_test, y_pred))
print("Recall:", metrics.recall_score(y_test, y_pred))
```
